```ruby
$:.unshift File.expand_path("..")

require 'market.rb'
require 'script/helpers.rb'
require 'daru/view'

Daru::View.plotting_library = :highcharts

# Rewriting and monkeypatching to remove the NMatrix requirement from nyaplot
# Daru makes its own NMatrix, but it doesn't have #to_nm.
class Array
  def mean
    ary = self.to_a.compact
    ary.inject(0) {|s, v| s + v } / ary.size.to_f
  end
end
```

```ruby
sim  = simulate :year => 2018..2019, :drop => -0.1
buys = sim.holding

buys = buys.map do |buy|
  roi_per_day = buy.changes_over(25).map do |day, roi|
    days = day.trading_days_from(buy)

    days == 0 ? 0 : roi.to_f / days
  end
  
  [buy, roi_per_day]
end.to_h
```

```ruby
opts = {:xAxis => {:type => "integer"},
        :yAxis => {:title => {:text => "ROI per day"}, :min => -1},
        :legend => {:layout => 'vertical',
                    :align => 'right',
                    :verticalAlign => 'middle'}}

plt = Daru::View::Plot.new
plt.chart.options = opts
plt.chart.series_data.shift # pop the first empty series off
buys.each {|buy, row| plt.add_series :data => row, :type => :line }

plt.show_in_iruby
```

```ruby
actual_rpds = sim.results.filter {|h| h[:hold] }.map {|h| h[:ROI].to_f / h[:hold] }
[actual_rpds.mean, actual_rpds.median]
```

```ruby
betters = buys.map {|b, rois| rois.any? {|x| x > actual_rpds.mean } }
betters.count(true) / betters.size.to_f
```

```ruby
roi = sim.results.filter {|h| h[:hold] }.map {|h| h[:ROI] }.median
hold = sim.results.filter {|h| h[:hold] }.map {|h| h[:hold] }.median
roi / hold
```
